```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

**Exercise 1**

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

**Exercise 2**

```python

x = np.linspace(0, 50, 500) #start 0, end 50, 500 no. in between.
y = np.exp(-x / 10) * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = e^{-x/10} \sin(x)$')

plt.xlim(0, 50)  # Set x-axis limits
plt.ylim(-1, 1)  # Set y-axis limits

plt.xticks(np.arange(0, 51, 10))  # x-axis from 0 to 50 with intervals of 10
plt.yticks(np.arange(-1, 1.25, 0.25))  # y-axis from -1 to 1 with intervals of 0.25

# Labels and title with custom font sizes and colors
plt.xlabel('X', fontsize=15, color='darkorange')
plt.ylabel('Y', fontsize=15, color='darkorange',rotation=0)
plt.title(r'Plot of $y = e^{-x/10} \sin(x)$', fontsize=20)

# Add grid with custom opacity
plt.grid(alpha=0.25)

# Display the plot
plt.show()
```
