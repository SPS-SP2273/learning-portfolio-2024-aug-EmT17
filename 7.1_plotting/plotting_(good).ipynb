```python


# Generate 100 points between -π and π
x = np.linspace(-np.pi, np.pi, num=100)  
sin_x = np.sin(x)

# Create a figure with subplots
fig, ax = plt.subplots(2, 3)  # Adjust the layout as needed

# Plotting sin_x on the correct subplot with matching x and y lengths
ax[1, 2].plot(x, sin_x, 'r')  # This assumes ax[1, 2] is a valid subplot

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```
