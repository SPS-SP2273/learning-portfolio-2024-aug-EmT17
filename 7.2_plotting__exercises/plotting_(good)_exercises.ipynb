```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

**Exercie 1**

```python
#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True, sharey='row')

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label=r'$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r', linestyle=':', label=r'$\cos^2 x$')
axes[0, 0].set_title(r'$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel(r'Cosine Value')
axes[0, 0].set_ylim(-1.25, 1.25)
axes[0, 0].set_xlim(-1.25, 1.25)
axes[0, 0].fill_between(x, cos_x, -1.25, color='r', alpha=.125) 
axes[0, 0].legend(loc=r'upper right')

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label=r'$\sin x$')
axes[0, 1].plot(x, sin_x**2, color='b', linestyle='-', label=r'$\sin^2 x$')
axes[0, 1].fill_between(x, sin_x, -2, color='g', alpha=.125)  # Corrected range
axes[0, 1].set_xlim(-1.25, 1.25)
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].set_title(r'$\sin x$ & $\sin^2x$')
axes[0, 1].legend(loc=r'upper right')  # Moved legend to top right

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label=r'$\sin 2x$')

# Highlight everything below the curve down to the bottom of the plot
axes[1, 0].fill_between(x, fun1_x, axes[1, 0].get_ylim()[0], color='b', alpha=.125)
axes[1, 0].set_title(r'$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Function Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'0', r'$\pi/2$', r'$\pi$'])
axes[1, 0].legend()

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color=r'y', label=r'$\sin 2x$')
axes[1, 1].fill_between(x, fun2_x, axes[1, 0].get_ylim()[0], color='y', alpha=.125)  # Corrected range
axes[1, 1].set_title(r'$e^{-x}\sin 2x$')
axes[1, 1].set_xlabel('Angle (radians)')
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].legend()

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.5)

# Adjust layout to remove y-axis padding
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)

# Automatically adjust subplot parameters to give specified padding
plt.tight_layout()

# save the figure
plt.savefig('After_Plot.jpg')
plt.show()
```
