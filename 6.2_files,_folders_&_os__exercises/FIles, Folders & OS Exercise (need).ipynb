**Task 4**

```python
def rename_my_file(old_filename):
    month_to_number = {
        'jan': '01', 'feb': '02', 'mar': '03', 'apr': '04',
        'may': '05', 'jun': '06', 'jul': '07', 'aug': '08',
        'sep': '09', 'oct': '10', 'nov': '11', 'dec': '12'}

    month_abbr, day_city = old_filename.split('-', 1)
    day, city_extension = day_city.split('_', 1)

    month_number = month_to_number[month_abbr]
    
    new_file_name = f"{month_number}-{day}-{city_extension}"
    return new_file_name

txt_files = glob.glob('os-collaboration-exercise-data/*.txt') #to get all the txt files

# Loop through each .txt file and rename it
for file in txt_files:
    
    old_filename = file.split('/', 1)
    temp = old_filename[1]
    
    old_filename = temp.lstrip()
    
    new_name = rename_my_file(old_filename)
    
    directory = f'os-collaboration-exercise-data'
    new_path = os.path.join(directory, new_name)
    
    os.rename(file, new_path)
    #show the renamed files 
    print(new_path)
```

```python
# Define the source directory and destination directory
source_directory = f'os-collaboration-exercise-data/'
destination_directory = f'os-collaboration-exercise-data/'

# Move files into corresponding folders
for file_name in os.listdir(source_directory):
    # Process only .txt files
    if file_name.endswith('.txt'):
        basename = file_name.split('.')[0]
        # Process only Cairo files, as required by task
        if 'cairo' in basename:
            city_name = basename.split('-')[-1]
        
            # Ensure the city name is in the correct format
            city_name = city_name.capitalize()
    
            # Construct old and new file paths
            old_file_path = os.path.join(source_directory, file_name)
            new_city_folder = os.path.join(destination_directory, city_name)
            new_file_path = os.path.join(new_city_folder, file_name)
    
            # Move the file to the new folder
            shutil.copy(old_file_path, new_file_path)
    
            # Print progress
            print(f"Copied {file_name} to {city_name} folder.")
```

**Task 7**

```python
# Define the source directory and destination directory
source_directory = f'os-collaboration-exercise-data/'
destination_directory = f'os-collaboration-exercise-data/'

# List of cities
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

# Create a folder for each city if it does not exist
for city in cities:
    city_folder = os.path.join(destination_directory, city)
    if not os.path.exists(city_folder):
        os.makedirs(city_folder)

# Copy files for each city into their respective folders
for city in cities:
    # Use glob to list all files related to the current city
    city_files = glob.glob(os.path.join(source_directory, f'*{city.lower()}*'))
    
    # Copy each city file to the respective city folder
    for file_path in city_files:
        # Extract the filename from the path
        file_name = os.path.basename(file_path)
        # The below if statement is to filter for the case where the file in question
        # is the city folder created in previous task
        if file_name == city:
            continue
        
        # Define the new path in the city folder
        new_file_path = os.path.join(destination_directory, city, file_name)
        print(file_name)
        
        # Copy the file to the new location
        shutil.move(file_path, new_file_path)
        
        # Print progress
        print(f"Moved {file_name} to {city} folder.")
```
